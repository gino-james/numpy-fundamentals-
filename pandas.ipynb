```python
# prompt: can you creat a histogram using sebonrn lib

# Assuming you have a DataFrame named 'df' with a column named 'SAT'
# If not, replace 'df' and 'SAT' with your actual DataFrame and column name

# Create a histogram using seaborn
sns.histplot(df['SAT'], bins=20, kde=True)  # kde=True adds a kernel density estimate curve

plt.xlabel('SAT Score')
plt.ylabel('Frequency')
plt.title('Histogram of SAT Scores')
plt.show()
```

```python
# prompt: create a Scatter Plots

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Assuming you have a DataFrame named 'df' with columns 'SAT' and 'Accept'
# If not, replace 'df', 'SAT', and 'Accept' with your actual DataFrame and column names

# Create a scatter plot using seaborn
sns.scatterplot(x='SAT', y='Accept', data=df, color='red')

plt.xlabel('SAT Score')
plt.ylabel('Acceptance Rate')
plt.title('Scatter Plot of SAT Score vs. Acceptance Rate')
plt.show()
```
